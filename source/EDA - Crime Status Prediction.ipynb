# 1. Data Pre-Processing

```python
# Check for any duplicate rows in the DataFrame 'cd'
duplicate_rows = cd[cd.duplicated()]

# If duplicate rows are found:
if not duplicate_rows.empty:
    # Print a message indicating that duplicate rows are found
    print("Duplicate rows found, removing duplicates...")

    cd.drop_duplicates(inplace=True)
```

```python
# Create a new column 'DateTimeValues'and converting the 'DATE OCC' column to datetime values
cd['DateTimeValues'] = pd.to_datetime(cd['DATE OCC'])
```

```python
# Extract the year from the 'DateTimeValues' column and create a new column 'year'
cd['year'] = cd['DateTimeValues'].dt.year

# Extract the month from the 'DateTimeValues' column and create a new column 'month'
cd['month'] = cd['DateTimeValues'].dt.month

# Extract the day from the 'DateTimeValues' column and create a new column 'day'
cd['day'] = cd['DateTimeValues'].dt.day
```

```python
cd['TIME OCC'] = cd['TIME OCC'].astype(str).str.zfill(4)
cd['HOUR OCC'] = cd['TIME OCC'].apply(lambda t: int(t[:2]))

# Extracting year, month, day and hour of the crime occured
```

```python
# dropping columns which are of similar information as other columns
cd = cd.drop(axis=1, columns=['Mocodes', 
                              'Rpt Dist No',
                              'Part 1-2',
                              'Crm Cd 1', # same as feature 'Crime Code', can be easily dropped
                              'Crm Cd 2', 
                              'Crm Cd 3', 
                              'Crm Cd 4',
                             'Date Rptd',
                             'Cross Street'])
```

```python
cd.info()
```

Column Name (Column 1): The name of the column in the DataFrame.

Data Type (Column 2): The data type of the values in the column (e.g., int64, object, float64, datetime64[ns])

```python
cd.isnull().sum()
```

```python
# Fill missing values in the 'Vict Sex' column with the value 'Unknown'
cd['Vict Sex'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Vict Descent' column with the value 'Unknown'
cd['Vict Descent'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Premis Cd' column with the value 'Unknown'
cd['Premis Cd'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Premis Desc' column with the value 'Unknown'
cd['Premis Desc'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Weapon Desc' column with the value 'Unknown'
cd['Weapon Desc'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Status' column with the value 'Unknown'
cd['Status'].fillna('Unknown', inplace=True)

# Fill missing values in the 'Weapon Used Cd' column with the value 0
cd['Weapon Used Cd'].fillna(0, inplace=True)
```

### 2.5 Number of Crimes by Sex

```python
cd['Vict Sex'].replace(to_replace=['H', '-', 'N'], value='X', inplace = True)
```

```python
def no_crime_by_sex(cd):
    # Group the DataFrame by 'Vict Sex' and count the occurrences of each group
    no_crime_by_sex = pd.DataFrame(cd.groupby('Vict Sex')['Vict Descent'].count()).reset_index()

    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(no_crime_by_sex['Vict Descent'], labels=no_crime_by_sex['Vict Sex'], autopct='%1.1f%%', startangle=90)
    plt.title('Percentage of Crimes By Sex')
    plt.show()
    
no_crime_by_sex(cd)
```

We can draw the conclusion that the crimes happening with male are comparatively higher than the female.

### 2.6 Distribution of Victim Age with Histogram

```python
plt.figure(figsize=(28, 4))

sns.set(style="whitegrid")

# Create a displot with a histogram and KDE for the 'Vict Age' column in the DataFrame 'cd'
sns.displot(cd['Vict Age'], kde=True, color='skyblue', bins=20, height=6)
plt.xlabel('Victim Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Victim Age with Histogram and KDE', fontsize=14)

# Add a vertical line for the mean age
mean_age = cd['Vict Age'].mean()
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean Age: {mean_age:.2f}')
plt.legend()

# Add a text annotation for the mean age
plt.text(mean_age + 1, plt.gca().get_ylim()[1] * 0.9, f'Mean Age: {mean_age:.2f}', color='red', fontsize=10)
plt.show()
```

### 2.7 Top Most weapons used in the crimes

```python
weapon_counts = cd['Weapon Desc'].value_counts().nlargest(11)[1:]

colors = sns.color_palette("pastel", len(weapon_counts))

# Create a horizontal bar plot for the top 10 (excluding the 1st which is 'unknown' i.e null values) weapon descriptions
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=weapon_counts.values, y=weapon_counts.index, palette=colors)

# Add labels and title
plt.xlabel('Frequency', fontsize=18)
plt.ylabel('Weapon Description', fontsize=18)
plt.title('Top 10 Weapon Descriptions', fontsize=20)

# Display the frequency values on the bars
for i, v in enumerate(weapon_counts.values):
    ax.text(v + 0.2, i, str(v), color='black', va='center', fontsize=10)

plt.gca().invert_yaxis()
plt.show()
```

The person who is committing the crime is using their bodily force in most of the cases followed by abusing the victim verbally

### 2.8 Visualization of Status Description

```python
status_counts = cd['Status Desc'].value_counts()

colors = sns.color_palette("viridis", len(status_counts))

# Create a horizontal bar plot for the status descriptions using Seaborn
plt.figure(figsize=(9, 6))
ax = sns.barplot(x=status_counts.values, y=status_counts.index, palette=colors)

# Add labels and title
plt.xlabel('Frequency')
plt.ylabel('Status Description')
plt.title('Status Descriptions')

# Display the frequency values on the bars
for i, v in enumerate(status_counts.values):
    ax.text(v + 0.2, i, str(v), color='black', va='center', fontsize=10)

plt.gca().invert_yaxis()
plt.show()
```

### 2.9 Distribution of Crime ccurrences based on Time Intervals

```python
# Define custom time intervals
time_bins = [6, 12, 16, 21, 5]

cd['Time Interval'] = pd.cut(cd['HOUR OCC'], bins=sorted(time_bins), labels=['6AM-12PM', '12PM-4PM', '4PM-9PM', '9PM-6AM'])

# Grouping by 'Time Interval' and counting occurrences
grouped_data = cd['Time Interval'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=grouped_data.index, y=grouped_data.values)
plt.title('Distribution of Occurrences based on Time Intervals')
plt.xlabel('Time Intervals')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

### 2.10 Number of Crimes happened in each month

```python
# Map numerical months to their abbreviated names
month_names = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
               7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

# Convert the 'month' column to categorical with custom month names
cd['month'] = cd['month'].map(month_names)
plt.figure(figsize=(15, 8))

# Create a count plot using seaborn to visualize the distribution of months
ax = sns.countplot(x="month", data=cd,
                   facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("dark", 12))  # Assuming 12 months

plt.xticks(rotation=45, fontsize=14)
plt.yticks(rotation=0, fontsize=14)
ax.set_xlabel("Month", fontsize=14)
ax.set_ylabel("Count", fontsize=14)
ax.set_title("Distribution of Crimes by Month", fontsize=16)
plt.show()
```

The resulting count plot visualizes the distribution of occurrences across different months. As we can see there is not much difference in the crimes as per months.
