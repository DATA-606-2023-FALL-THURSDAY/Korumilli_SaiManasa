```python
plt.figure(figsize=(20,20))
plt.style.use('seaborn-whitegrid')
sns.heatmap(cd.corr(), annot=True, cmap='RdYlGn')
plt.tight_layout()
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.fontsize=20
plt.title('Correlation Matrix')
plt.show()
```

```python
# Check for any duplicate rows
duplicate_rows = cd[cd.duplicated()]
if not duplicate_rows.empty:
    print("Duplicate rows found, removing duplicates...")
    cd.drop_duplicates(inplace=True)
```

```python
# dropping columns with a majority of missing values
cd = cd.drop(axis=1, columns=['Mocodes', 
                              'Crm Cd 1', # same as feature 'Crime Code', can be easily dropped
                              'Crm Cd 2', 
                              'Crm Cd 3', 
                              'Crm Cd 4',
                             'Date Rptd'])
```

```python
cd['DateTimeValues'] = pd.to_datetime(cd['DATE OCC'])
```

```python
cd['year'] = cd['DateTimeValues'].dt.year
cd['month'] = cd['DateTimeValues'].dt.month
cd['day'] = cd['DateTimeValues'].dt.day
```

```python
cd['TIME OCC'] = cd['TIME OCC'].astype(str).str.zfill(4)
cd['HOUR OCC'] = cd['TIME OCC'].apply(lambda t: int(t[:2]))
```
